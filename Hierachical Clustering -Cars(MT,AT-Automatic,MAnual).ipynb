```python
# Import Libraries

import numpy as np
import pandas as pd
import scipy 
from scipy.cluster.hierarchy import dendrogram ,linkage
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
from pylab import rcParams
import seaborn as sb
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import AgglomerativeClustering
import sklearn.metrics as sm
import os
```

```python
os.getcwd()
```

```python
# Import Data Set

os.chdir('C:\\Users\\HP\\Downloads\\Batch 93 Day34\\')
```

```python
cars=pd.read_csv('mtcars.csv')
display(cars)
```

```python
# Create X

x=cars.iloc[:,[1,3,4,6]].values
print(x)
```

```python
# Create Y

y=cars.iloc[:,9].values
print(y)
```

```python
# Plot Graph

plt.scatter(x[y==0,0],x[y==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],s=100,c='blue',label='Cluster 2')
plt.legend()
plt.show()
```

```python
# Create the Linkage

z=linkage(x,'ward')
display(z)
```

```python
# Create Dendrogram

dendrogram(z,truncate_mode='lastp',p=12,leaf_rotation=45,leaf_font_size=15,show_contracted=True)
plt.title('Truncated Hierarchical Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')
plt.axhline(y=500)
plt.axhline(y=150)
plt.show()
```

```python
# Create Model with Ward

k=2
hclustering=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
y_hc=hclustering.fit_predict(x)
display(sm.accuracy_score(y,y_hc))
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='r',label='Cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='g',label='Cluster 2')
plt.legend()
plt.show()
```

```python
# two Cluster with Linkage - Single

z=linkage(x,'single')
dendrogram(z,truncate_mode='lastp',p=12,leaf_rotation=45.,leaf_font_size=15,show_contracted=True)
plt.title('Truncated Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()
k=2
hclustering=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='single')
y_hc=hclustering.fit_predict(x)
display(sm.accuracy_score(y,y_hc))
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='blue',label='Cluster 2')
plt.legend()
plt.show()
```

```python
# two Cluster with Linkage - Complete

z=linkage(x,'complete')
dendrogram(z,truncate_mode='lastp',p=12,leaf_rotation=45.,leaf_font_size=15,show_contracted=True)
plt.title('Truncated Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()
k=2
hclustering=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='complete')
y_hc=hclustering.fit_predict(x)
display(sm.accuracy_score(y,y_hc))
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='blue',label='Cluster 2')
plt.legend()
plt.show()
```

```python
# two Cluster with Linkage - Average

z=linkage(x,'average')
dendrogram(z,truncate_mode='lastp',p=12,leaf_rotation=45.,leaf_font_size=15,show_contracted=True)
plt.title('Truncated Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()
k=2
hclustering=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='average')
y_hc=hclustering.fit_predict(x)
display(sm.accuracy_score(y,y_hc))
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='blue',label='Cluster 2')
plt.legend()
plt.show()
```

```python
# Concat and Save data in Excel

Final=pd.concat([cars,pd.DataFrame(y_hc,columns=['Cluster Number'])],axis=1)
Final.to_excel('Hierarchical Clustering- Cars(MT,AT-Automatic.MAnual).xlsx')
display(Final)
```
